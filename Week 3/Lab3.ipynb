Step 1) Translation(Shifting an image)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('labphoto.jpg')
print(image.shape)
translation_matrix = np.float32([[1, 0, -100], [0,1, -50]])
translated_image = cv2.warpAffine (image,
translation_matrix, (image.shape[1],
image.shape [0]))
print(translated_image.shape)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.axis('off')
plt.show()
```

```python
print(image.shape)
translation_matrix = np.float32([[1, 0, 100], [0,1, 50]])
translated_image = cv2.warpAffine (image,
translation_matrix, (image.shape[1],
image.shape [0]))
print(translated_image.shape)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.axis('off')
plt.show()
```

Step 2) Rotation

```python
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center,
45, 1.0)
rotated_image = cv2.warpAffine (image,
rotation_matrix, (w, h))
plt.imshow(cv2.cvtColor (rotated_image,
cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (45 degrees) ')
plt.axis('off')
plt.show()
```

Step 3) Scaling

```python
Factor = 2
scaling_matrix = cv2.getRotationMatrix2D((0,0), 0, Factor)
scaled_image = cv2.warpAffine (image,
scaling_matrix, (w*Factor, h*Factor))
plt.imshow(cv2.cvtColor(scaled_image,
cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (2x) ')
plt.axis('off')
plt.show ()
print (scaled_image.shape)
```

```python
Factor = 0.5
scaling_matrix = cv2.getRotationMatrix2D((0,0), 0, Factor)
scaled_image = cv2.warpAffine(image,
scaling_matrix, (int(w*Factor), int(h*Factor)))
plt.imshow(cv2.cvtColor(scaled_image,
cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (0.5x) ')
plt.axis('off')
plt.show ()
print (scaled_image.shape)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('labphoto.jpg')
(h, w) = image.shape[:2]

Factor = 10
scaling_matrix = cv2.getRotationMatrix2D((0,0), 0, Factor)
scaled_image = cv2.warpAffine(image,
scaling_matrix, (int(w*Factor), int(h*Factor)))
plt.imshow(cv2.cvtColor(scaled_image,
cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (10x) ')
plt.axis('off')
plt.show ()
print (scaled_image.shape)
```

Step 4) Shearing

```python
m = 0.67
shearing_matrix = np.float32([[1, m, 0], [0, 1,
0]])
sheared_image = cv2.warpAffine(image,
shearing_matrix, (w, h))
plt.imshow(cv2.cvtColor(sheared_image,
cv2.COLOR_BGR2RGB))
plt.title('Sheared Image (x-axis, factor ' + str(m) + ')')
plt.axis('off')
plt.show ()
```
