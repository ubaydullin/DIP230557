```python
from PIL import Image

# 1. Open the image
img_pil = Image.open("/content/images/IMG_2279.jpg")

# 2. Print image metadata
print(f"Type: {type(img_pil)}")
print(f"Size: {img_pil.size}")  # (Width, Height)
print(f"Mode: {img_pil.mode}")  # e.g., RGB

# 3. Display the image
img_pil.show()
```

```python
import numpy as np

# Convert PIL image to NumPy array
img_np = np.array(img_pil)

# Print array shape and data type
print(f"Array Shape: {img_np.shape}") # (Height, Width, Channels)
print(f"Data Type: {img_np.dtype}")
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# 1. Read the image
img_plt = mpimg.imread("/content/images/IMG_2279.jpg")

# 2. Print shape
print(f"Shape: {img_plt.shape}")

# 3. Display it
plt.imshow(img_plt)
plt.axis('off') # Optional: hides the axes
plt.show()
```

```python
from skimage import io, img_as_float
import matplotlib.pyplot as plt

# 1. Read the image
img_sk = io.imread("/content/images/IMG_2279.jpg")

# 2. Convert to float
img_float = img_as_float(img_sk)

# 3. Print min/max pixel values
print(f"Min value: {img_float.min()}")
print(f"Max value: {img_float.max()}")

# 4. Display
io.imshow(img_float)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# 1. Read in grayscale and color
img_bgr = cv2.imread("/content/images/IMG_2279.jpg")
img_gray = cv2.imread("/content/images/IMG_2279.jpg", cv2.IMREAD_GRAYSCALE)

# 2. Convert BGR to RGB (OpenCV default is BGR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# 3. Display using matplotlib
plt.imshow(img_rgb)
plt.title("Converted RGB")
plt.show()
```

```python
# Perform Canny edge detection
edges = cv2.Canny(img_gray, 100, 200)

# Display the result
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```
