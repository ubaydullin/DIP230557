# Task 1
Step 1: Subtraction (Darkening)

```python
import cv2
import matplotlib.pyplot as plt

# Load and convert to RGB
image = cv2.imread('/content/pexels-photo-359620.jpeg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Split the channels
R, G, B = cv2.split(image_RGB)

# Subtract 100 from Red channel only (Task 1, Question 3)
R_sub = cv2.subtract(R, 100)

# Merge back
red_darkened_image = cv2.merge([R_sub, G, B])

# Display
plt.imshow(red_darkened_image)
plt.title('Red Channel Subtracted (-100)')
plt.axis('off')
plt.show()
```

# Step 2 Addition (Brightening)

```python
# Add 100 to Red channel only (Task 2, Question 6)
R_add = cv2.add(R, 100)
red_brightened_image = cv2.merge([R_add, G, B])

plt.imshow(red_brightened_image)
plt.title('Red Channel Added (+100)')
plt.axis('off')
plt.show()
```

Step 3: Multiplication (Contrast)

```python
# Task 3: Experiment with factors (Question 1)
high_contrast = cv2.multiply(image_RGB, 2.0) # Multiply by 2.0
low_contrast = cv2.multiply(image_RGB, 0.5)  # Multiply by 0.5

# Displaying low contrast example
plt.imshow(low_contrast)
plt.title('Multiplied by 0.5 (Lower Contrast)')
plt.show()
```

# Step 4: Division (Brightness Reduction)

```python
# Task 4: Experiment with division (Question 1)
divided_image = cv2.divide(image_RGB, 3.0)

plt.imshow(divided_image)
plt.title('Divided by 3.0 (Darker)')
plt.show()
```

# Task 2

```python
import cv2
import matplotlib.pyplot as plt

# 1. Read the image
image = cv2.imread('/content/pexels-photo-359620.jpeg')

# 2. Convert from BGR (OpenCV default) to RGB for Matplotlib
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 3. Split the image into R, G, and B channels
# In RGB format, index 0 is Red, 1 is Green, 2 is Blue
R, G, B = cv2.split(image_RGB)

# 4. Increase intensity of the Red channel only by 100
# Using cv2.add ensures values stay capped at 255 (saturate)
R_increased = cv2.add(R, 100)

# 5. Merge the modified Red channel back with the original G and B channels
modified_image = cv2.merge([R_increased, G, B])

# 6. Display original and modified images for comparison
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(modified_image)
plt.title('Red Channel +100')
plt.axis('off')

plt.show()
```
